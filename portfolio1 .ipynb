```python
# Portfolio 1 - Expressions in Python
# Simple Grocery Budget Calculator
# Created by: 
# ==========================================

# Item prices
bread_price = 45
milk_price = 68
noodles_price = 20

# Quantity of items
bread_qty = 2
milk_qty = 1
noodles_qty = 5

# ------------------------------------------
# MULTIPLICATION
# Compute total cost per item
# ------------------------------------------

bread_total = bread_price * bread_qty
milk_total = milk_price * milk_qty
noodles_total = noodles_price * noodles_qty

# ------------------------------------------
# ADDITION
# Compute overall grocery total
# ------------------------------------------

grocery_total = bread_total + milk_total + noodles_total

# Display grocery expenses
print("GROCERY SUMMARY")
print("Bread Total:", bread_total)
print("Milk Total:", milk_total)
print("Noodles Total:", noodles_total)
print("Overall Total:", grocery_total)

# ------------------------------------------
# SUBTRACTION
# Student budget after buying groceries
# ------------------------------------------

student_budget = 500
remaining_budget = student_budget - grocery_total

print("\nRemaining Budget:", remaining_budget)

# ------------------------------------------
# DIVISION
# Average spending per item type
# ------------------------------------------

average_spending = grocery_total / 3

print("Average Spending:", average_spending)

# ------------------------------------------
# FLOOR DIVISION
# How many full groups of 50 pesos
# ------------------------------------------

full_groups = grocery_total // 50

print("Groups of 50 Pesos:", full_groups)

# ------------------------------------------
# MODULO / REMAINDER
# Remaining pesos after grouping by 50
# ------------------------------------------

remaining_pesos = grocery_total % 50

print("Remaining Pesos:", remaining_pesos)

# ------------------------------------------
# EXPONENT / POWER
# Example of exponentiation
# ------------------------------------------

sample_power = 2 ** 4

print("2 raised to 4 is:", sample_power)

# ------------------------------------------
# OPERATOR PRECEDENCE
# Python follows PEMDAS
# ------------------------------------------

expression_result = 5 + 2 * 3 ** 2

print("Result of 5 + 2 * 3 ** 2:", expression_result)
```
