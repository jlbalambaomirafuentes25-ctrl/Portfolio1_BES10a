```python
"""
Portfolio 2: Conditionals
Simple Budget Status Checker
Created by:
This program checks the user's remaining budget
and gives a status using conditional statements.
"""

# Input section
budget = float(input("Enter your total budget: "))
expenses = float(input("Enter your total expenses: "))

# Computation
remaining = budget - expenses

# Output section with conditionals
print("\n--- Budget Status ---")

if remaining < 0:
    print("Status: Over budget! You spent more than your budget.")
elif remaining <= 500:
    print("Status: Almost empty budget. Be careful with spending.")
else:
    print("Status: Healthy budget. You're managing well!")

# Final output
print("Remaining money:", remaining)
print("Finish program.")
```
