```python
"""
Portfolio 3: Functions
Student Grade Checker

This program checks the student's grade
using functions and conditional statements.
"""

# Function definition
def check_grade(grade):

    if grade >= 90:
        return "With Honor"

    elif grade >= 75:
        return "Passed"

    else:
        return "Failed"


# Input section
name = input("Enter student name: ")
grade = float(input("Enter student grade: "))

# Function call
result = check_grade(grade)

# Output
print("\n--- Result ---")
print("Student:", name)
print("Grade:", grade)
print("Status:", result)

print("Program finished.")
print (result)
```
