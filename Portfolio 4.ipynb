```python
"""
Portfolio 4: Iteration


Daily Expense Tracker

This program asks the user to enter daily expenses.
It uses loops to compute the total and average expense.
"""

# Variables
total = 0
count = 0

print("Enter your expenses.")
print("Type 0 to stop.\n")

# While loop
while True:

    expense = float(input("Enter expense: "))

    # Break statement
    if expense == 0:
        break

    # Add values
    total = total + expense
    count = count + 1

# Output
print("\n--- Expense Summary ---")
print("Number of expenses:", count)
print("Total expense:", total)

# Conditional statement
if count > 0:
    average = total / count
    print("Average expense:", average)

print("Program finished.")
```
